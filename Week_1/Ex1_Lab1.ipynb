```python
import pandas as pd
df=pd.read_csv("mtcars.csv")
df
```

```python
#Best mpg
bestmpg = df['mpg'].max() 
df[df['mpg']==bestmpg]
```

```python
#Worst mpg
worstmpg = df['mpg'].min()
df[df['mpg']==worstmpg]
```

```python
#Best Horsepower
besthp = df['hp'].max()
df[df['hp']==besthp]
```

```python
import numpy as np
quartiles = np.percentile(df['disp'], [25, 50, 75])
print('Min: %.3f' % df['disp'].min())
print('Q1: %.3f' % quartiles[0])
print('Median: %.3f' % quartiles[1])
print('Q3: %.3f' % quartiles[2])
print('Max: %.3f' % df['disp'].max())
```

```python
#Median Horsepower
print("Median horsepower is %f" % df['hp'].median())
```

```python
print(f"Mileage for Manual cars {df['mpg'][df['am'] == 0].mean()}")
print(f"Mileage for Automatic cars {df['mpg'][df['am'] == 1].mean()}")
```
