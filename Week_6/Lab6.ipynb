## User-Based Similarity

#### Read the “ratings.csv” file and create a pivot table with index=‘userId’, columns=‘movieId’,values = “rating

```python
#Importing the datset

df = pd.read_csv("ratings.csv")
df
```

```python
df_pivot = pd.pivot_table(df,values = "rating", index = "userId", columns = "movieId")
df_pivot
```

```python
df_pivot.fillna(value=0, inplace=True)
df_pivot
```

#### sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of users. pairwise_distances() takes a metric parameter for what distance measure to use. Use cosine similarity for finding similarity among users. Use the following packages.
#### from sklearn.metrics import pairwise_distances
#### from scipy.spatial.distance import cosine, correlation

```python
distance=pairwise_distances(df_pivot,metric="cosine")
distance
```

```python
x = pd.DataFrame(np.squeeze(distance))
x
```

### Find the 5 most similar user for user with user Id 10.

```python
users_movie = pairwise_distances(df_pivot, metric='cosine')
user_similarity = users_movie[10 - 1, :]
five_similar = user_similarity.argsort()
five_similar = five_similar[:5]
print(f"Most Similar Users to User ID 10 are: {[x+1 for x in five_similar]} in the same high to low order")
```

#### 7. Use the “movies” dataset to find out the names of movies, user 2 and user 338 have watched in common and how they have rated each one of them.

```python
#Importing the datset

movies = pd.read_csv("movies.csv")
movies
comm = pd.DataFrame(pd.merge(df, movies, on="movieId"))
comm
```

```python
user2_movie = pd.DataFrame(comm[comm["userId"] == 2][["title","rating"]])
user2_movie.head(30)
```

```python
user338_movie = pd.DataFrame(comm[comm["userId"] == 338][["title","rating"]])
user338_movie.head(40)
```

```python
similar = pd.merge(user2_movie, user338_movie, on="title")
similar.head()
```

#### Use the movies dataset to find out the common movie names between user 2 and user 338 with least rating of 4.0

```python
similar[(similar["rating_x"] >= 4) & ((similar["rating_y"] >= 4))]
```
